### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
### This task will be done in following steps:
### 1. First get the webpage https://www.naukri.com/
### 2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the location” field.
### 3. Then click the search button.
### 4. Then scrape the data for the first 10 jobs results you get.
### 5. Finally create a dataframe of the scraped data

```python
url='https://www.naukri.com/'
driver.get(url)
search_job=driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div/div[1]/div/div/div/input") # finding web element for job search bar
search_job.send_keys("Data Analyst") # Write on search bar
search_locn=driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input") # finding web element for location
search_locn.send_keys("Bangalore") # write on location
search_btn=driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div/div[6]") # finding web element for search button
search_btn
search_btn.click() # clicking on the search button
```

```python
job_titlesA=[]
title_tagA=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]') # extracting job titles
for i in title_tagA:
    job_titlesA.append(i.text)
```

```python
company_namesA=[]
company_tagsA=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]') # extracting company names
for i in company_tagsA:
    company_namesA.append(i.text)
```

```python
experienceA=[]
exp_tagsA=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]') # extrcting experience
for i in exp_tagsA:
    experienceA.append(i.text)
```

```python
locationA=[]
loc_tagsA=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in loc_tagsA:
    locationA.append(i.text)
```

```python
jobsA=pd.DataFrame()
jobsA['Job title']=job_titlesA[:10]
jobsA['Company']=company_namesA[:10]
jobsA['Experience']=experienceA[:10]
jobsA['Location']=locationA[:10]
jobsA
```
