### Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Creating empty lists
brand_names=[]
product_name=[]
price=[]
return_exchange=[]
exp_delivery=[]
availability=[]
product_url=[]

start=0
end=3
for page in range(start,end): # for loop for scraping 3 pages
    urls=driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]/a')
    for i in urls:
        product_url.append(i.get_attribute('href'))
    nxt_button=driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')#scraping the list of buttons from the page
    try:
        driver.get(nxt_button.get_attribute('href'))#getting the link from the list for next page
    except:
        driver.get(nxt_button.get_attribute('href'))    
    
        
```

```python
for product in product_url:
    driver.get(product)
    try:
        pr_name=driver.find_elements_by_xpath('//span[@id="productTitle"]')
        for i in pr_name:
            product_name.append(i.text)
        
    except :
        product_name.append('-')
        
    try:
        brand=driver.find_elements_by_xpath('//table[@class="a-normal a-spacing-micro"]/tbody/tr[1]/td[2]/span[1]')
        for i in brand:
            brand_names.append(i.text)
    except :
        brand_names.append('-')
        
    try:
        pricetag=driver.find_elements_by_xpath('//span[@class="a-price aok-align-center"]/span/span[2]')
        for i in pricetag:
            price.append(i.text)
            
    except :
        price.append('-')
        
    try:
        deliverytag=driver.find_elements_by_xpath('//div[@id="ddmDeliveryMessage"]/b')
        for i in deliverytag:
            exp_delivery.append(i.text)
        
    except :
        exp_delivery.append('-')
        
    try:
        avaltag=driver.find_elements_by_xpath('//div[@id="availability"]/span')
        for i in avaltag:
            availability.append(i.text)
            
    except :
        availability.append('-')
        
    try:
        returntag=driver.find_elements_by_xpath('//div[@id="RETURNS_POLICY"]/span/div/a')
        for i in returntag:
            return_exchange.append(i.text)
            
    except :
        return_exchange.append('-')
        
        
            
```

```python
df=pd.DataFrame({'Brand':brand_names,'Product Name':product_name,'Price':price,'Delivery':exp_delivery,'Availability':availability,'Return/Exchange':return_exchange,'Product URL':product_url})
df
```
